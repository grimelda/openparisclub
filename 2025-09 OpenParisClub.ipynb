```python
#@markdown F U N C T I O N S

#@markdown read(url, queries)
def read(
        url,
        queries,
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'},
        ):
    '''
    url: string e.g. "https://clubdeparis.org/en/traitements/bulgaria-17-04-1991/en"
    queries: dict of labels + html divs
    '''

    query_data = {}
    query_data['url'] = url

    result = requests.get(url, headers=headers)
    soup = BeautifulSoup(result.content, "html.parser")

    for query, id in queries.items():
        # find string
        try:
            if query == 'Attached file':
                datum = soup.find(class_ = id).find('a', href=True)['href']
                datum = 'https://clubdeparis.org'+datum
            else:
                datum = soup.find(
                                'div',
                                attrs={'class': id}
                            ).get_text()
            # 1: remove query label from result
            datum = datum.replace(query, '')
            datum = datum.replace('\xa0','')

            # 3: remove unwanted character combinations
            datum = datum.replace('\n','; ')
            datum = datum.replace('\s\t','')
            datum = datum.replace(' ;',';')
            datum = datum.strip(' .;\t')

            # 2: substitute double characters
            datum = datum.replace('\n\n', '; ')
            datum = datum.replace('\t\t', '\t')
            datum = datum.replace('\s\s', ' ')
            datum = datum.replace(';;', ';')
            datum = datum.replace('.;', '.')

            query_data[query] = datum

        except AttributeError:
                pass

    return query_data
```
